```python
import chromadb

client = chromadb.Client()
collection = client.create_collection(name="my_collection")
```

```python
collection.add(
    documents=[
        "This document is about messi",
        "This document is about neymar"
    ],
    ids = ['id1', 'id2']
)
```

```python
collection.delete(ids= ['id1', 'id2','id3','id4'])
```

```python
all_docs = collection.get()
all_docs
```

```python
results = collection.query(
    query_texts=['soccer'],
    n_results=2
)
results
```

```python
collection.delete(ids=all_docs['ids'])
collection.get()
```
